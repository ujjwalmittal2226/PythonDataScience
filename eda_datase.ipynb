```python
df['wi'] = df['Weight'] + df['Life Span']
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#plt.xlim(35)

# Set y-axis limits (e.g., exercise requirement range from 0 to 3)
#plt.ylim(1.5, 3)
#plt.scatter(df['Weight'], df['Exercise Requirements (hrs/day)'])  # Assuming 'Exercise_Requirements' column exists
plt.scatter(df['Life Span'], df['Weight'])
plt.xlabel('Life Span')
plt.ylabel('Weight')
plt.title('Weight vs. Exercise Requirements')
plt.grid(True)
plt.show()
```

```python

sns.histplot(df['Weight'], kde=True)  # Added kde for kernel density estimation
plt.xlabel('Life Span')
plt.ylabel('Frequency')
plt.title('Distribution of Dog Weights')
plt.show()
```

```python
df.groupby('Origin')['Weight'].mean().plot(kind='bar')
plt.xlabel('Origin')
df.sort_values('Origin')
plt.ylabel('Average Weight (kg)')
plt.title('Average Weight by Origin')
plt.xticks(rotation=135) # Rotate x-axis labels for better readability
plt.xlim(0,15)
plt.show()
```

```python
sns.boxplot(x='Origin', y='Weight', data=df)
plt.xlabel('Origin')
plt.ylabel('Weight (kg)')
plt.title('Weight Distribution by Origin')
plt.xticks(rotation=45)
plt.xlim(0,15)
plt.show()
```

```python
plt.plot(df['Life Span'], df['Weight'])  # Assuming 'Life Span' column exists
plt.xlabel('Life Span')
plt.ylabel('Weight (kg)')
plt.title('Weight vs. Life Span')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x='Weight', y='Exercise Requirements (hrs/day)', data=df)
plt.xlabel('Weight (kg)')
plt.ylabel('Exercise Requirements')
plt.title('Weight vs. Exercise Requirements')
plt.xlim(25)
plt.ylim(3)
plt.show()
```

```python
df['Origin'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribution of Dog Origins')

plt.show()
```

```python
sns.histplot(df['Weight'], kde=True)
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.title('Distribution of Dog Weights')
plt.show()
```

```python
sns.barplot(x='Origin', y='Weight', data=df, ci=None)  # ci=None removes confidence intervals
plt.xlabel('Origin')
plt.ylabel('Average Weight (kg)')
plt.title('Average Weight by Origin')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='Origin', y='Weight', data=df)
plt.xlabel('Origin')
plt.ylabel('Weight (kg)')
plt.title('Weight Distribution by Origin')
plt.xticks(rotation=45)
plt.show()
```
